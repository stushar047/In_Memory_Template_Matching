```python
full_image = cv2.imread("PeanutBottle1/img_0.jpg",0) # shape (360,320) 

temp_image = cv2.imread("template/template2.jpg",0) # shape (360,320)

temp_row , temp_col = temp_image.shape

Data_template = temp_image.reshape(-1,)

row_stride=20

col_stride =20

Corr_template_soft=np.zeros((iter_row+1, iter_col+1))

iter_row, iter_col = ((full_image.shape[0]-temp_row)//row_stride), (full_image.shape[1]-temp_col)//col_stride

for ir in range(iter_row+1):

    for ic in range(iter_col+1):

        Data_sample = full_image[ir*col_stride:ir*col_stride+temp_row, ic*col_stride:ic*col_stride+temp_col].reshape(-1,)

        Corr_template_soft[ir,ic]=soft_pcc(Data_template, Data_sample)

def soft_pcc(Data_template, Data_sample):

    Data_1 = Data_template - np.mean(Data_template)

    Data_2 = Data_sample - np.mean(Data_sample)

    return np.sum(Data_1* Data_2)/(np.sqrt(sum(Data_1**2))*np.sqrt(sum(Data_2**2)))
```

```python
import numpy as np

# Constants for the dimensions of the array
ROWS, COLS = 64, 128

# Creating a 2D array with 8-bit unsigned integer values (0-255)
array = np.random.randint(0, 256, size=(ROWS, COLS), dtype=np.uint8)

# Function to write the 2D array in C++ style to a file
def write_2d_array_to_file(filename, array):
    with open(filename, 'w') as file:
        file.write("const uint8_t array[{}][{}] = {{\n".format(array.shape[0], array.shape[1]))
        for i, row in enumerate(array):
            file.write("    {")
            file.write(", ".join(map(str, row)))
            file.write("},\n")
        file.write("};\n")

# Write the array to 'array.txt'
write_2d_array_to_file('temp.txt', array)
```

```python
def ADC (input_value, min_input = -0.150, max_input = 0.150, bits=16):

    # Calculate ratio
    ratio = (input_value - min_input) / (max_input - min_input)

    # Scale ratio to output range
    output_value = int(ratio * 2**bits - 2**(bits-1))

    return output_value
```
